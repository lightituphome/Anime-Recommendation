```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

Data Collection and preprocessing

```python
data=pd.read_csv('anime.csv')
```

```python
selected_features = ['genre','type','episodes','rating','members']
print(selected_features)
```

```python
for feature in selected_features:
  data[feature]=data[feature].fillna('')
```

```python
combined_features=data['genre']+data['type']
```

```python
print(combined_features)
```

```python
vectorizer = TfidfVectorizer()
```

```python
feature_vectors=vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

Cosine Similarity

```python
similarity=cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
print(similarity.shape)
```

Anime Recommendation System

```python
anime_name=input('Enter the anime name:')
```

```python
list_of_all_titles=data['name'].tolist()
print(list_of_all_titles)
```

```python
find_close_match=difflib.get_close_matches(anime_name,list_of_all_titles)
print(find_close_match)
```

```python
close_match=find_close_match[0]
print(close_match)
```

```python
index_of_the_anime=data[data.name==close_match]['anime_id'].values[0]
print(index_of_the_anime)
```

```python
similarity_score=list(enumerate(similarity[index_of_the_anime]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
sorted_similar_anime=sorted(similarity_score,key=lambda x:x[1],reverse=True)
print(sorted_similar_anime)
```

```python
print('Anime suggested for you: \n')

i=1

for anime in sorted_similar_anime:
  index=anime[0]
  title_from_index=data[data.index==index]['name'].values[0]
  if(i<30):
    print(i,'.',title_from_index)
    i+=1
```
